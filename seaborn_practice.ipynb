```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#in inside seaborn datasests checking
sns.get_dataset_names()
```

```python
#creating data varaiable or dataset and checking top 5 head values
penguins=sns.load_dataset("penguins")
penguins.head()
```

```python
#Unique values count in species and island column 
penguins["species"].value_counts()
penguins["island"].value_counts()
```

```python
#creating scatterplot diagram and in seaborn different baground styles is there (themes)
#white,whitegrid,dark,darkgrid,ticks
sns.set_style("whitegrid")
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island');
```

```python
#how to change figure size like dimensions width and length
plt.figure(figsize=(10,6))
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island');
```

```python
#Here, we are performing with numeric data with numeric data
#How to change context of the plot like changing plot elememts looks (labels,lines,ticks,legends)
#Here total types are : paper,notebook(default),poster,talk
sns.set_context("poster")
plt.figure(figsize=(8,5))
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island');
```

```python
#Here, we are performing with alpha data with numeric data
#using palettes : it is like adding colur to the plot (set1,set2,set3,viridis,cubehelix...,etc)
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.scatterplot(data=penguins,x='species',y='body_mass_g',hue='island',style="sex",alpha=1,palette='cubehelix');
```

```python
penguins.head(1)
```

```python
#stripplot using dodge,jitter (dodge : it will separate and give in group by ) (jitter : it will separate dots in visible way like separtion )
sns.stripplot(data=penguins,x='species',y='body_mass_g',hue="island",palette='Set1',jitter=True,dodge=True);
```

```python
#swarm plot ( it will avoid overlapping, clearly it can give )
sns.set_context("paper")
sns.swarmplot(data=penguins,x='species',y='body_mass_g',hue='island',palette='bright');
sns.despine()
```

```python
#Histogram plot (counts of values grouped into bins)
plt.figure(figsize=(8,5))
sns.histplot(data=penguins,x='body_mass_g',bins=10,hue='sex',multiple='stack');
```

```python
#regression plot (shows a scatterplot along with a best-fit regression line to visualize the relationship between two variables.)
sns.regplot(data=penguins,x='body_mass_g',y='bill_length_mm',color='green',scatter=False);
```

```python
#line plot (a graph that shows how a value changes over time or across an ordered sequence.)
plt.figure(figsize=(15,5))
sns.set_style("whitegrid")
sns.set_context("paper")
sns.lineplot(data=penguins,x='body_mass_g',y='bill_length_mm',hue='sex',style='island');
```

```python
#jointplot (the relationship between two variables along with their individual distributions.)
sns.jointplot(data=penguins,x='body_mass_g',y='bill_length_mm',kind='scatter',hue='sex');
#if we will give with kind parameter we can change middle main plot (our opinion, we can change plot )
```

```python
#Bar plot (By default it will give avg(mean) values it will give for each categories)
sns.barplot(data=penguins,y='body_mass_g',x='species',hue='sex',palette=["red","pink"],estimator=np.var);
```

```python
#countplot (a plot that shows how many times each category appears (counts of categories))
sns.set_style("darkgrid")
sns.set_context("notebook")
sns.countplot(data=penguins,x='species',hue='island',palette='Set1');
```

```python
#Box plot (a graph that shows the distribution of data based on percentiles  including the min, 25th percentile,median,75th percentile & max  and highlights outliers.)
sns.boxplot(data=penguins,x='species',y='body_mass_g',hue='sex');
```

```python
#violin plot (same like boxplot but here it will give with full details in a distrubtion shaped)
sns.violinplot(data=penguins,x='species',y='body_mass_g',hue='sex',split=True,inner='box');
```

```python
#violin plot (same like boxplot but here it will give with full details in a distrubtion shaped)
sns.violinplot(data=penguins,x='species',y='body_mass_g',hue='sex',split=True);
```

```python
#swarm plot (shows all dots clearly without overlapping.)
#sns.violinplot(data=penguins,x='species',y='body_mass_g');
sns.swarmplot(data=penguins,x='species',y='body_mass_g',hue='sex',size=5,palette='Set2');
```

```python
#KDE: KERNAL DENSITY ESTIMATE plot (shows where values are high or low using a smooth curve.)
sns.kdeplot(data=penguins,x='body_mass_g',hue='species',fill=True);
```

```python
#RUG PLOT (**Rug plot is used to show data distribution along one axis only by marking each data point with a small line)
sns.rugplot(data=penguins,x='body_mass_g',hue='sex',height=0.5,palette='Set1');
```

```python
#PAIR PLOT (shows scatterplots and histograms or kde for every pair of numeric variables in a dataset, helping you see relationships easily)
sns.pairplot(data=penguins,hue='sex',diag_kind='kde');
```

```python
#PAIR GRID (a customizable grid that lets you create your own pairplot with different plots in each section)
grid=sns.PairGrid(data=penguins,hue='sex',palette='dark');
grid.map_upper(sns.scatterplot) #upperplot
grid.map_lower(sns.kdeplot) #lowerplot
grid.map_diag(sns.histplot) #diagonal 
```
